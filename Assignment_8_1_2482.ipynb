```python
# Task Description #1 (Password Strength Validator – Apply AI in Security Context)

import string
def is_strong_password(password):
    if len(password) < 8:
        return False
    if " " in password:
        return False
    has_upper = False
    has_lower = False
    has_digit = False
    has_special = False
    special_chars = string.punctuation
    for ch in password:
        if ch.isupper():
            has_upper = True
        elif ch.islower():
            has_lower = True
        elif ch.isdigit():
            has_digit = True
        elif ch in special_chars:
            has_special = True
    # The return statement should be outside the loop to ensure all characters are checked
    return has_upper and has_lower and has_digit and has_special
assert is_strong_password("Abcd@123") == True
assert is_strong_password("abcd123") == False
assert is_strong_password("ABCD@1234") == False
assert is_strong_password("Abc 123@") == False
assert is_strong_password("Strong@9") == True
print("All password strength test cases passed")
```

```python
#Task Description #2 (Number Classification with Loops – Apply AI for Edge Case Handling)

def classify_number(n):
    if not isinstance(n, (int, float)) or n is None:
        return "Invalid input"
    for _ in range(1):
        if n > 0:
            return "Positive"
        elif n < 0:
            return "Negative"
        else:
            return "Zero"
assert classify_number(10) == "Positive"
assert classify_number(-5) == "Negative"
assert classify_number(0) == "Zero"
assert classify_number(1) == "Positive"
assert classify_number(-1) == "Negative"
assert classify_number("100") == "Invalid input"
assert classify_number(None) == "Invalid input"
assert classify_number(3.5) == "Positive"
print("All number classification test cases passed")
```

```python
# Task Description #3 (Anagram Checker – Apply AI for String Analysis)

import string
def is_anagram(str1, str2):
    if not isinstance(str1, str) or not isinstance(str2, str):
        return False
    clean1 = ""
    for ch in str1.lower():
        if ch.isalnum():
            clean1 += ch

    clean2 = ""
    for ch in str2.lower():
        if ch.isalnum():
            clean2 += ch

    return sorted(clean1) == sorted(clean2)
assert is_anagram("listen", "silent") == True
assert is_anagram("hello", "world") == False
assert is_anagram("Dormitory", "Dirty Room") == True
assert is_anagram("", "") == True
assert is_anagram("Triangle", "Integral") == True
assert is_anagram("Apple", "Pabble") == False
assert is_anagram("A gentleman", "Elegant man!") == True
print("All anagram test cases passed")
```

```python
# Task Description #4 (Inventory Class – Apply AI to Simulate Real-World Inventory System)

class Inventory:
    def __init__(self):
        self.stock = {}

    def add_item(self, name, quantity):
        if quantity <= 0:
            return "Invalid quantity"
        self.stock[name] = self.stock.get(name, 0) + quantity

    def remove_item(self, name, quantity):
        if name not in self.stock:
            return "Item not found"
        if quantity <= 0 or quantity > self.stock[name]:
            return "Invalid quantity"
        self.stock[name] -= quantity

    def get_stock(self, name):
        return self.stock.get(name, 0)

# Test cases
inv = Inventory()
inv.add_item("Pen", 10)
assert inv.get_stock("Pen") == 10
inv.remove_item("Pen", 5)
assert inv.get_stock("Pen") == 5
inv.add_item("Book", 3)
assert inv.get_stock("Book") == 3
inv.remove_item("Pen", 10)  # This should not remove any as only 5 are left and remove_item handles invalid quantity
assert inv.get_stock("Pen") == 5
inv.add_item("Pen", 5)
assert inv.get_stock("Pen") == 10
assert inv.get_stock("Pencil") == 0
print("All inventory test cases passed")
```

```python
# Task Description #5 (Date Validation & Formatting – Apply AI for Data Validation)

from datetime import datetime

def validate_and_format_date(date_str):
    try:

        date_obj = datetime.strptime(date_str, "%m/%d/%Y")

        return date_obj.strftime("%Y-%m-%d")
    except (ValueError, TypeError):
        return "Invalid Date"
assert validate_and_format_date("10/15/2023") == "2023-10-15"
assert validate_and_format_date("02/30/2023") == "Invalid Date"
assert validate_and_format_date("01/01/2024") == "2024-01-01"


assert validate_and_format_date("13/01/2023") == "Invalid Date"
assert validate_and_format_date("2/5/2023") == "2023-02-05"
assert validate_and_format_date(None) == "Invalid Date"
assert validate_and_format_date("12-25-2023") == "Invalid Date"

print("All date validation test cases passed ")
```
