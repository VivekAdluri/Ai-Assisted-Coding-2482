#Task-02

### Gemini AI Generated Code

```python
# Sample list of user records
users = [
    {"name": "Alice", "age": 30, "city": "New York"},
    {"name": "Bob", "age": 24, "city": "Los Angeles"},
    {"name": "Charlie", "age": 35, "city": "Chicago"},
    {"name": "David", "age": 24, "city": "Houston"}
]

print("Original list:")
for user in users:
    print(user)

# Sort the list of dictionaries by 'age' in-place
users.sort(key=lambda user: user["age"])

print("\nSorted list (by age - Gemini AI):")
for user in users:
    print(user)
```

This code uses the `list.sort()` method, which sorts the list in-place. The `key` argument is a `lambda` function that tells `sort()` to use the value associated with the 'age' key for comparison during sorting. This modifies the original `users` list directly.

### Cursor AI Generated Code

```python
# Sample list of user records (re-initialized for independent sorting demonstration)
users_cursor = [
    {"name": "Alice", "age": 30, "city": "New York"},
    {"name": "Bob", "age": 24, "city": "Los Angeles"},
    {"name": "Charlie", "age": 35, "city": "Chicago"},
    {"name": "David", "age": 24, "city": "Houston"}
]

print("Original list:")
for user in users_cursor:
    print(user)

# Sort the list of dictionaries by 'age' and create a new sorted list
sorted_users_cursor = sorted(users_cursor, key=lambda user: user["age"])

print("\nSorted list (by age - Cursor AI):")
for user in sorted_users_cursor:
    print(user)
```

This code uses the built-in `sorted()` function, which takes an iterable (like a list) and returns a *new* sorted list, leaving the original list unchanged. Similar to `list.sort()`, the `key` argument with a `lambda` function specifies sorting by the 'age' key.

### Comparison of Clarity and Performance

-   **Readability:** Both approaches are highly readable and beginner-friendly. The use of `key=lambda user: user["age"]` clearly indicates the sorting criterion. The `sorted()` function might be marginally more intuitive for beginners as it explicitly creates a new sorted list, which can prevent accidental modification of the original data.

-   **Simplicity:** Both methods are simple Pythonic ways to sort. `list.sort()` is simple for in-place modification, while `sorted()` is simple for obtaining a new sorted list without altering the original. The choice depends on whether the original list needs to be preserved.

-   **Performance for small datasets:** For small datasets like the one demonstrated, the performance difference between `list.sort()` and `sorted()` is negligible. Both use Python's Timsort algorithm, which is highly optimized. The primary performance consideration is that `sorted()` creates a new list, which incurs a small overhead for memory allocation and copying, whereas `list.sort()` modifies in-place, thus avoiding this overhead. However, for small datasets, this difference is practically insignificant.

```python
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Error: Cannot divide by zero!"
    return x / y

def calculator():
    print("\n--- Simple Python Calculator ---")
    print("Select operation:")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    print("5. Exit")

    while True:
        choice = input("Enter choice(1/2/3/4/5): ")

        if choice in ('1', '2', '3', '4'):
            try:
                num1 = float(input("Enter first number: "))
                num2 = float(input("Enter second number: "))
            except ValueError:
                print("Invalid input. Please enter numbers only!")
                continue

            if choice == '1':
                print(num1, "+", num2, "=", add(num1, num2))
            elif choice == '2':
                print(num1, "-", num2, "=", subtract(num1, num2))
            elif choice == '3':
                print(num1, "*", num2, "=", multiply(num1, num2))
            elif choice == '4':
                print(num1, "/", num2, "=", divide(num1, num2))
        elif choice == '5':
            print("Exiting calculator. Goodbye!")
            break
        else:
            print("Invalid input. Please enter a valid choice (1-5).")

calculator()
```

```python
# Version 1 (Gemini) - Armstrong Number Program

def is_armstrong_number(number_str):
    """Checks if a given number (as a string) is an Armstrong number."""
    try:
        number = int(number_str)
    except ValueError:
        return "Error: Invalid input. Please enter an integer."

    if number < 0:
        return "Error: Please enter a non-negative number."

    # Convert number to string to find its length (number of digits)
    num_str = str(number)
    n = len(num_str)

    sum_of_powers = 0
    temp_number = number

    while temp_number > 0:
        digit = temp_number % 10
        sum_of_powers += digit ** n
        temp_number //= 10

    if sum_of_powers == number:
        return f"{number} is an Armstrong number."
    else:
        return f"{number} is not an Armstrong number."

def run_armstrong_checker():
    """Runs the Armstrong number checker program."""
    print("\n--- Armstrong Number Checker ---")
    user_input = input("Enter a non-negative integer: ")
    result = is_armstrong_number(user_input)
    print(result)

# To run the Armstrong checker, uncomment the line below:
# run_armstrong_checker()

# Explanation
"""
This program defines two functions: `is_armstrong_number` and `run_armstrong_checker`.

- `is_armstrong_number(number_str)`:
    - Takes a string input, attempts to convert it to an integer, and handles `ValueError` for non-integer inputs.
    - Validates for non-negative numbers.
    - Calculates the number of digits (`n`) in the input number.
    - Iterates through each digit of the number, raising each digit to the power of `n` and summing these results.
    - Compares the final sum with the original number. If they are equal, it's an Armstrong number; otherwise, it's not.

- `run_armstrong_checker()`:
    - This function provides the user interface for the program.
    - It prompts the user to enter a non-negative integer.
    - Calls `is_armstrong_number()` with the user's input.
    - Prints the result returned by `is_armstrong_number()`.

The logic is encapsulated in `is_armstrong_number` for reusability, and `run_armstrong_checker` handles user interaction.

# Sample Output
--- Armstrong Number Checker ---
Enter a non-negative integer: 153
153 is an Armstrong number.

--- Armstrong Number Checker ---
Enter a non-negative integer: 9
9 is an Armstrong number.

--- Armstrong Number Checker ---
Enter a non-negative integer: 123
123 is not an Armstrong number.

--- Armstrong Number Checker ---
Enter a non-negative integer: -5
Error: Please enter a non-negative number.

--- Armstrong Number Checker ---
Enter a non-negative integer: abc
Error: Invalid input. Please enter an integer.
"""
```
